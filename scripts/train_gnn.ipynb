```python
import sys, os
ROOT = os.path.abspath(os.path.join(os.getcwd(), "..")) 
sys.path.append(os.path.join(ROOT, "src"))
import matplotlib.pyplot as plt 
from dagc.sparsifiers.gnn.train import train_gnn_sparsifier_on_dataset
```

```python
# ckpt_path = "results/checkpoints/gnn_keep0.6_ic_diffusion.pt"
ckpt_path = "results/checkpoints/gnn_keep0.8_rw_loss_more_node_features_1_step_4_layers.pt"
model, history = train_gnn_sparsifier_on_dataset(
                    train_dir="../data/synthetic_graphs/ba/train", 
                    val_dir="../data/synthetic_graphs/ba/val",
                    keep_ratio=0.8,
                    num_steps=1, 
                    num_epochs=15, 
                    learning_rate=1e-3, 
                    lambda_sparsity=1.5,
                    print_every=1, 
                    checkpoint_path=ckpt_path, 
                    )
epochs = history["epoch"] 
plt.figure()
plt.plot(epochs, history["train_loss"], label="train") 
plt.plot(epochs, history["val_loss"], label="val") 
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("RW + sparsity loss over epochs")
plt.show()
```
